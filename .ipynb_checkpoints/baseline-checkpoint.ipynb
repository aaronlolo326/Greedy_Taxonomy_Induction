# Utility

```python
def pickledata(filename,data):
    pickling_on = open(filename,"wb")
    pickle.dump(data, pickling_on)
    pickling_on.close()

def loadpickleddata(filename):
    with open(filename, "rb") as input_file:
        return pickle.load(input_file)

def dd():
    return defaultdict(int)

def normalize_dict(d):
    m = max(list(d.values()))
    norm_d = {k: v/m for k, v in d.items()}
    return norm_d

def get_normFreqDiff_table(h2H,terms):
    NFD = defaultdict(lambda: defaultdict(float))
    for term1 in terms:
        for term2 in terms:
            NFD[term1][term2] = normalized_freq_diff(h2H,term1,term2)
    return NFD

def sorted_key(d):
    sorted_key = sorted(d,key=d.get)
    return sorted_key

# normalize_dict({'a':1, 'b':2, 'c':3})
```

# GloVe

```python
# load the Stanford GloVe model
def get_w2v(filename):
    model = KeyedVectors.load_word2vec_format(filename, binary=False)
    return model

def get_cos_sim(term1,term2):

    words1 = term1.split(" ")
    words2 = term2.split(" ")
    assign_mean = False
    try:
        _ = [model.get_vector(word) for word in words1]
        simwords1 = words1
    except:
        try: #find vec of first's last word
            _ = model.get_vector(words1[-1])
            simwords1 = [words1[-1]]
        except:
            assign_mean = True
    try:
        _ = [model.get_vector(word) for word in words2]
        simwords2 = words2
    except:
        try: #find vec of first's last word
            _ = model.get_vector(words2[-1])
            simwords2 = [words2[-1]]
        except:
            assign_mean = True

            
    if assign_mean == False:
        try: #find vec of first's and second's words
            sim = model.n_similarity(simwords1,simwords2)
        except Exception as e: #at least one word missing
            print (e)
        return sim
    else:
        return 0

def get_all_cosSim(gs_terms):
    cosSim_dict = {term1: {term2:get_cos_sim(term1,term2) for term2 in gs_terms} for term1 in gs_terms }
    return cosSim_dict

def get_terms_avg_cosSim(gs_terms):
    terms_avg_cosSim = np.mean([get_cos_sim(term1,gs_terms[idx2]) for idx1,term1 in enumerate(gs_terms) for idx2 in range(idx1+1,len(gs_terms))])
    return terms_avg_cosSim  
            
```

# Features

```python
def Endswith(x, y):
    return int(y.endswith(x))
def Contains(x, y):
    return int(x in y)
def Suffix_match(x, y):
    k = 7
    for i in range(k):
        if x[-i - 1:] != y[-i - 1:]:
            return i
    return k
def LCS(x, y):
    match = SequenceMatcher(None, x, y).find_longest_match(0, len(x), 0, len(y))
    res = 2.0 * match.size / (len(x) + len(y))  # [0, 1]
    return int(round(res, 1) * 10)  # [0,10]
def LD(x, y):
    res = 2.0 * (len(x) - len(y)) / (len(x) + len(y))  # (-2,2)
    return int(round(res, 1) * 10 + 20)  # [0, 40]

# def doubly_normalized_freq_diff(H2h,h2H, x, y):
#     if x not in h2H or y not in h2H[x] or h2H[x][y] == 0:
#         a = 0
#     else:
#         denom = sum([h2H[x][H]/max(list(H2h[H].values())) for H in h2H[x]])
#         a = float(h2H[x][y])/max(list(H2h[y].values())) / denom
#     if y not in h2H or x not in h2H[y] or h2H[y][x] == 0:
#         b = 0
#     else:
#         denom = sum([h2H[y][H]/max(list(H2h[H].values())) for H in h2H[y]])
#         b = float(h2H[y][x])/max(list(H2h[x].values())) / denom
#     res = a - b  # [-1, 1]
#     return a,b
#     #return int(res * 10) + 10  # [0, 20]

def normalized_freq_diff(hypo2hyper, x, y):
    if x not in hypo2hyper or y not in hypo2hyper[x] or hypo2hyper[x][y] == 0:
        a = 0
    else:
        a = float(hypo2hyper[x][y]) / max(hypo2hyper[x].values())
    if y not in hypo2hyper or x not in hypo2hyper[y] or hypo2hyper[y][x] == 0:
        b = 0
    else:
        b = float(hypo2hyper[y][x]) / max(hypo2hyper[y].values())
    res = a - b  # [-1, 1]
    return res
    #return int(res * 10) + 10  # [0, 20]

def generality(hyper2hypo, x):
    if x not in hyper2hypo:
        b = 0
    else:
        b = np.log(1 + len([i for i in hyper2hypo[x] if hyper2hypo[x][i] != 0]))
    return b

def generality_diff(hyper2hypo, x, y):
    if x not in hyper2hypo:
        b = 0
    else:
        b = np.log(1 + len([i for i in hyper2hypo[x] if hyper2hypo[x][i] != 0]))
    if y not in hyper2hypo:
        a = 0
    else:
        a = np.log(1 + len([i for i in hyper2hypo[y] if hyper2hypo[y][i] != 0]))
    res = a - b  # (-7.03, 7.02)
    # return res
    return int(res) + 7  # [0, 14]

def get_all_features(x, y, sub_feat=False, hypo2hyper=None, hyper2hypo=None, lower2original=None):
    feat = {}
    #feat['Capitalization'] = Capitalization(x, y, lower2original)
    feat['Endswith'] = Endswith(x, y)
    feat['Contains'] = Contains(x, y)
    feat['Suffix_match'] = Suffix_match(x, y)
    feat['LCS'] = LCS(x, y)
    feat['LD'] = LD(x, y)
    if sub_feat:
#         if hypo2hyper is None or hyper2hypo is None:
#             print ('features.py: hypo2hyper not loaded')
#             exit(-2)
#         feat['Freq_diff'] = normalized_freq_diff(hypo2hyper, x, y)
        feat['General_diff'] = generality_diff(hyper2hypo, x, y)
    return feat
```

# Precounted Hearst Counts

```python

def get_hearst_cnts():
    H2h = defaultdict(dd)
    h2H = defaultdict(dd)
    line_cnt = 0
    with open(hearst_cnt_file,encoding = 'utf-8') as f:
        for line in f:
            H2h[line.strip().split('\t')[1].replace("_"," ")][line.strip().split('\t')[0].replace("_"," ")]=int(line.strip().split('\t')[2])
    for hyper in H2h:
        for hypo in H2h[hyper]:
            h2H[hypo][hyper] = H2h[hyper][hypo]
    return H2h,h2H

```

# Taxonomy Class

```python
class taxonomy:
    
    def __init__(self, gs_terms):
        self.G = nx.DiGraph()
        self.G.add_nodes_from(gs_terms)
        self.gs_terms = gs_terms
        
    def load_taxo_file(self, taxo_file):
        self.gs_taxo = nx.DiGraph()
        self.gs_taxo.add_nodes_from(self.gs_terms)
        #print ("self gs terms: ",self.gs_terms)
        #print ("self gs nodes: ",self.gs_taxo.nodes)
        with open(taxo_file,'r',encoding = 'utf-8') as f:
            for line in f:
                #print (line)
                index, term1, term2 = line.strip("\n").split("\t")
                self.gs_taxo.add_edge(term2.lower(),term1.lower())
        #print ("self gs nodes after edges: ",self.gs_taxo.nodes)
    def term_from_taxo_file(self, taxo_file):
        term_set = set()
        with open(taxo_file,'r',encoding = 'utf-8') as f:
            for line in f:
                index, term1, term2 = line.strip("\n").split("\t")
                term_set.add(term1.lower())
                term_set.add(term2.lower())
        return term_set
    
    def save_gs_taxo(self, gs_taxo):
        self.gs_taxo = gs_taxo
        
    def connect_substring(self):
        for term1 in self.G.nodes:
            for term2 in self.G.nodes:
                if term1.endswith(term2) and (not term1 == term2) and self.G.in_degree(term1) ==0:
                    #print (term1,self.G.in_degree(term1))
                    self.G.add_edge(term2,term1)
        #print (self.G.in_degree('antarctic ocean'))
    
    def get_0indeg_nodes(self):
        nodes = [key[0] for key in self.G.in_degree() if key[1] == 0]
        #print (nodes)
        return nodes
    
    
    def get_linkable_nodes(self,cand_node):
        UG = self.G.to_undirected()
        nodes = set(self.G.nodes).difference(nx.node_connected_component(UG, cand_node)).difference(set(cand_node))
        return nodes
    
    def connect2nodes(self,parent,child):
        self.G.add_edge(parent,child)
    
    def evaluate(self):
        
        gs_anc = {node: nx.descendants(self.gs_taxo,node) for node in self.gs_taxo.nodes}
        sys_anc = {node: nx.descendants(self.G,node) for node in self.G.nodes}
        #print (self.gs_taxo.nodes)
        common_anc_cnt = sum([len(gs_anc[node].intersection(sys_anc[node])) for node in self.gs_taxo.nodes])
        sys_anc_cnt = sum([len(sys_anc[node]) for node in self.G.nodes])
        gs_anc_cnt = sum([len(gs_anc[node]) for node in self.gs_taxo.nodes])
        ancestor_prec = common_anc_cnt/sys_anc_cnt
        ancestor_recall = common_anc_cnt/gs_anc_cnt
        ancestor_f1 = 2*((ancestor_prec*ancestor_recall)/(ancestor_prec+ancestor_recall))
        
        gs_edges = set(self.gs_taxo.edges)
        #print (gs_edges)
        sys_edges = set(self.G.edges)
        #print (sys_edges)
        common_edge_cnt = len(gs_edges.intersection(sys_edges))
        sys_edge_cnt = len(sys_edges)
        gs_edge_cnt = len(gs_edges)
        edge_prec = common_edge_cnt/sys_edge_cnt
        edge_recall = common_edge_cnt/gs_edge_cnt
        try:
            edge_f1 = 2*((edge_prec*edge_recall)/(edge_prec+edge_recall))
        except:
            edge_f1 = 0
        
        if(math.isclose(edge_prec,edge_recall) and math.isclose(edge_recall,edge_f1)):
            return ancestor_prec,ancestor_recall,ancestor_f1,edge_f1
        else:
            print ("hv bug")
            return ancestor_prec,ancestor_recall,ancestor_f1,edge_f1
```

# SLQS Class

```python
class slqs:
    
    def __init__(self,taxo_name,gs_terms):
        code_path = "../slqs_onlygsterms_{}".format(taxo_name)
        self.term2idx = loadpickleddata("{}/term2idx.pickle".format(code_path))
        self.median_ent = loadpickleddata("{}/top{}_median_ent.pickle".format(code_path,50))
        self.gs_terms = gs_terms
        self.sorted_gen_dict = self.sort_term_gen()
        
    def sort_term_gen(self):
        gen_dict = {}
        for term in self.gs_terms:
            try:
                gen_dict[term] = self.median_ent[self.term2idx[term]]
            except:
                gen_dict[term] = 0
        sorted_gen_key = sorted(gen_dict,key=gen_dict.get)
        sorted_gen_dict = {term:gen_dict[term] for term in sorted_gen_key}
        #print ("slqs_dict: ",sorted_gen_dict)
        return sorted_gen_dict
```

# Generate Taxonomy
* print edge connected one by one
* return a taxonomy object

```python
%load_ext line_profiler


def genTaxo(taxo_name,gs_terms,cosSim_dict,terms_avg_cosSim,hearst_NFD,taxoRL_NFD,NFD_thresold,slqs_gen_dict,
            taxorl_gen_dict,hearst_gen_dict,config):
    #gs_terms = ['ocean','anatartic ocean','arctic ocean','environment','marine life']
    len_gs_terms = len(gs_terms)
    
    taxo = taxonomy(gs_terms)
    
    if config['substring_connection']:
        taxo.connect_substring()
    
    slqs_coef = 1 if config['gen_measure']=='slqs' else 0
    hearst_coef = 1 if config['gen_measure']=='hearst' else 0
    taxoRL_coef = 1 if config['gen_measure']=='taxoRL' else 0
    if config['gen_measure']=='taxorl_hearst':
        hearst_coef,taxoRL_coef = 1,1
    if config['gen_measure']=='all':
        slqs_coef,hearst_coef,taxoRL_coef = 1,1,1 
        
    NFD_coef = 1 if config['connect_measure']=='NFD' else 0
    cosSim_coef = 1 if config['connect_measure']=='cosSim' else 0
    if config['connect_measure']=='all':
        NFD_coef,cosSim_coef = 1,1 
    
    #print (slqs_gen_dict)
    avg_gen_dict = {k: slqs_gen_dict[k]*slqs_coef + taxorl_gen_dict[k]*taxoRL_coef + hearst_gen_dict[k]*hearst_coef for k in slqs_gen_dict}
    #print (avg_gen_dict)
    avg_gen_key = sorted_key(avg_gen_dict)
    #print (len(avg_gen_key))
    
    while len(taxo.G.edges) < len_gs_terms-1:
        
        frontier = taxo.get_0indeg_nodes()
        spec_frontier = list(filter(lambda x: x in frontier,avg_gen_key))[:int((len(frontier)+1)/2)]
        
        #max_avg_cosSim = None
        max_avg_NFD = None
        NFD_tobe_link = None
        
        max_combined_measure = -3 # (-1,1) and (-1,1)
        combined_tobe_link = None
        max_combined_measure_split = None
        
        used_measure = None
        avg_cosSim = 0
        avg_NFD = 0
        
        for spec_term in spec_frontier:
                        
            linkable_nodes = taxo.get_linkable_nodes(spec_term)
            #print ("spec_term:",spec_term, ";linkable_nodes: ", linkable_nodes)
            if config['connect_measure'] == 'NFD' or config['connect_measure'] == 'all':
                spec_term_descendants = nx.descendants(taxo.G,spec_term)
                spec_term_descendants.add(spec_term)
                
            #if (spec_term == 'dynamic systems'):
             #   print (linkable_nodes,spec_frontier,spec_term_descendants)
                
            for linkable_node in linkable_nodes:
                if config['connect_measure'] == 'NFD' or config['connect_measure'] == 'all':
                #determine by NFD between spec&spec_descendent and linkable node
                    #linkable_node_ancestors = nx.ancestors(linkable_node)
                    avg_NFD = np.mean([taxoRL_NFD[spec_term_des][linkable_node] for spec_term_des in spec_term_descendants])
                                #+ np.mean([hearst_NFD[spec_term_des][linkable_node] for spec_term_des in spec_term_descendants])
                    #print (tobe_link,descendants, max_avg_NFD)
                if config['connect_measure'] == 'cosSim' or config['connect_measure'] == 'all':
                #determine by cosSim between spec 
                    linkable_node_children = list(taxo.G.successors(linkable_node)) #add own node trytry
                    linkable_node_children.append(linkable_node)
                    avg_cosSim = np.mean(list(map(lambda x:cosSim_dict[spec_term][x], linkable_node_children)))
                    if math.isnan(avg_cosSim):
                        avg_cosSim = terms_avg_cosSim
                        
                #if (spec_term == 'dynamic systems'):
                 #   pass#print ({spec_term_des: taxoRL_NFD[spec_term_des][linkable_node] for spec_term_des in spec_term_descendants})

                        
                if max_avg_NFD==None or avg_NFD >= max_avg_NFD:
                    max_avg_NFD = avg_NFD
                    NFD_tobe_link = (linkable_node,spec_term)
                
                if (max_avg_NFD <= NFD_thresold):
                    combined_measure = NFD_coef * avg_NFD + cosSim_coef * avg_cosSim
                    if combined_measure >= max_combined_measure:
                        max_combined_measure = combined_measure if not math.isnan(combined_measure) else -3
                        combined_tobe_link = (linkable_node,spec_term)
                        max_combined_measure_split = (avg_NFD,avg_cosSim)
# separate measure codes:
#                 if max_avg_cosSim==None or avg_cosSim >= max_avg_cosSim:
#                     max_avg_cosSim = avg_cosSim if not math.isnan(avg_cosSim) else -1
#                     tobe_link = (linkable_node,spec_term)
#                     children = linkable_node_children
#                 if max_avg_NFD==None or avg_NFD >= max_avg_NFD:
#                         max_avg_NFD = avg_NFD if not math.isnan(avg_NFD) else -1
#                         tobe_link = (linkable_node,spec_term)
#                         descendants = spec_term_descendants
        if max_avg_NFD <= NFD_thresold:
            tobe_link = combined_tobe_link
            used_measure = max_combined_measure_split
        else:
            tobe_link = NFD_tobe_link
            used_measure = max_avg_NFD
            
        print (len(taxo.G.edges),tobe_link,"NFD,(cosSim): ",used_measure)
        taxo.connect2nodes(tobe_link[0],tobe_link[1])
    
    return taxo
```

# Main

  * extract terms from .terms file
  * extract gold standard taxonomy from .taxo file
  * prepare cosSim of the terms
  * prepare normalized frequency different matrix for each term pairs, from taxoRL and hearst
  * prepare generality dictionary for each term, using taxoRL(NFD), hearst(NFD) and slqs
  * for each config, generate taxonomy and record result log and pickle taxonomy data

```python

for taxo_name in taxo_names:
    terms_file = terms_dir+taxo_name+".terms"
    with open(terms_file, 'r', encoding = 'utf-8') as f:
        gs_terms = [line.strip().split('\t')[1].lower() for line in f]
    gs_taxo_file = gs_dir+taxo_name+".taxo"
    
    cosSim_dict = get_all_cosSim(gs_terms)
    terms_avg_cosSim = get_terms_avg_cosSim(gs_terms)
    taxoRL_H2h,taxoRL_h2H = get_TaxoRL_cnts(taxo_name)
    hearst_H2h,hearst_h2H = get_hearst_cnts()
    taxoRL_NFD = get_normFreqDiff_table(taxoRL_h2H,gs_terms)
    hearst_NFD = get_normFreqDiff_table(hearst_h2H,gs_terms)
    NFD_thresold = 0.0
    slqs_gen_dict = normalize_dict(slqs(taxo_name,gs_terms).sorted_gen_dict)
    taxorl_gen_dict = normalize_dict({term: generality(taxoRL_H2h,term) for term in gs_terms})
    #taxorl_gen = sorted(taxorl_gen_dict,key=taxorl_gen_dict.get)
    hearst_gen_dict = normalize_dict({term: generality(hearst_H2h,term) for term in gs_terms})  
    #hearst_gen = sorted(hearst_gen_dict,key=hearst_gen_dict.get)
    for config in configs:
#         if(config['substring_connection']==False):
        taxo = genTaxo(taxo_name,gs_terms,cosSim_dict,terms_avg_cosSim,hearst_NFD,taxoRL_NFD,NFD_thresold,
                       slqs_gen_dict,taxorl_gen_dict,hearst_gen_dict,config)
        taxo.load_taxo_file(gs_taxo_file)
        ancestor_prec,ancestor_recall,ancestor_f1,edge_f1 = taxo.evaluate()
        with open(log_file,"a+") as f:
            f.write ("Done {} with {}\n".format(taxo_name,str(config)))
            f.write ("ancestor prec/recall/F1: {}, {}, {}; edge F1: {}\n".format(ancestor_prec,ancestor_recall,ancestor_f1,edge_f1))

        pickledata(pickled_digraph_file.format(taxo_name),taxo.G)
    
```
